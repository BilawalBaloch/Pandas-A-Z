# **Groupby**

```python
import pandas as pd
#marks of  Students
var = pd.DataFrame({"Name" : ["Ali","Ahmed","Bilawal","Yasir","Bilawal","Ali","Adnan","Imran","Rawal","Hamza"],
                      "Subject1" : [20,13,15,29,22,20,13,15,29,22],
                        "Subject2" : [23,33,25,29,12,22,23,22,21,21]})
var
```

making it as Groupby

```python
import pandas as pd
#marks of  Students
var = pd.DataFrame({"Name": ["Ali","Ahmed","Bilawal","Yasir","Bilawal","Ali","Adnan","Imran","Rawal","Hamza"],
                      "Subject1" : [20,13,15,29,22,20,13,15,29,22],
                        "Subject2" : [23,33,25,29,12,22,23,22,21,21]})
var
var_new =var.groupby ("Subject1")
var_new
for x,y in var_new:
   print(x)
   print(y)
```

picking single DATA

```python
import pandas as pd
#marks of  Students
var = pd.DataFrame({"Name": ["Ali","Ahmed","Bilawal","Yasir","Bilawal","Ali","Adnan","Imran","Rawal","Hamza"],
                      "Subject1" : [20,13,15,29,22,20,13,15,29,22],
                        "Subject2" : [23,33,25,29,12,22,23,22,21,21]})
var
var_new=var.groupby("Name").get_group("Bilawal")
var_new
```

# **Join and Append**

**Join**

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[5,6,7,8,9]},index=['a','b','c','d','e'])
var2 = pd.DataFrame({"R":[5,6,7,8,9],"E":[6,8,9,2,4]},index=['a','b','c','d','e'])
var1.join(var2)
```

```python
var1.join(var2,how="inner")
```

```python
var1.join(var2, how="outer")
```
